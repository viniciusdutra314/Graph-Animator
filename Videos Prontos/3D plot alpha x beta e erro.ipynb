```python
import matplotlib.pyplot as plt
import numpy as np
import mplcyberpunk
```

```python
n=100
x=np.linspace(0,2,n)
y=np.linspace(0,20,20*n)
X,Y=np.meshgrid(x,y)
def f(alpha,beta):
    y=[26.1,30.1,37.3,44.5,50.6,59.6,62.7,68.1]
    x=[10,20,30,40,50,60,70,80]
    erro=0
    for j in range(len(x)):
        erro+=(y[j]-x[j]*alpha-beta)**2
    return erro
Z=f(X,Y)
```

```python
ax=plt.axes(projection="3d")
ax.set_xlabel("Coeficiente Angular")
ax.set_ylabel("Termo independente")
ax.set_zlabel("Erro quadrÃ¡tico")
ax.scatter3D(0.63,18.9,2000,s=100,color="red",label="Menor erro")
plt.style.use('cyberpunk') 
ax.plot_wireframe(X,Y,Z,edgecolors="green",alpha=1,label="Superficie de Erros",lw=0.5)
plt.legend()
```
